```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/df_listo_para_segmentar.csv',sep=';')

df
```

```python
# Elimina las columnas 'Destino Turístico' y 'CUT Destino Turístico', 
df = df.drop(columns=['Destino Turistico', 'CUT Destino Turistico','Norte a Sur Origen', 'Norte a Sur Destino'])
```

```python
df.head()
```

```python
df = df.drop(columns=['Provincia Origen', 'Comuna Origen', 'CUT Comuna Origen', 'CUT Provincia Origen','CUT Provincia Destino', 'Provincia Destino'])
```

```python
# Renombrar las columnas
df = df.rename(columns={'CUT Temporada': 'Num Temporada', 'CUT Mes': 'Num Mes'})

df
```

```python
df.columns
```

```python
# Definir el nuevo orden de las columnas

column_order = ['CUT Region Origen', 'Region Origen', 'CUT Region Destino', 'Region Destino', 
                'CUT Comuna Destino', 'Comuna Destino', 'Anio', 'Num Mes', 'Mes', 
                'Num Temporada', 'Temporada', 'Viajes Ocasionales', 'PIB Region Origen', 
                'PIB Region Destino']

# Reordenar las columnas en el DataFrame
df = df[column_order]

df
```

```python
df_grouped = df.groupby(['Region Origen', 'Comuna Destino', 'Anio', 'Mes'], as_index=False).agg({
    'Viajes Ocasionales': 'sum',          # Sumar los viajes ocasionales
    'PIB Region Origen': 'first',         # Mantener el primer valor del PIB según la región de origen
    'PIB Region Destino': 'first',        # Mantener el primer valor del PIB según la región de destino
    'Temporada': 'first'                  # Tomar la temporada correspondiente según el mes y año
})

df_grouped
```

```python
# Hacemos el merge entre el df original y el df_grouped usando las columnas clave
df_merged = pd.merge(
    df, df_grouped, 
    on=['Region Origen', 'Comuna Destino', 'Anio', 'Mes'], 
    how='left',  # Usamos 'left' para mantener todas las filas de df original
    suffixes=('_original', '_grouped')  # Para distinguir las columnas duplicadas, si las hay
)

df_merged
```

```python
# Filtramos el DataFrame para obtener solo las filas correspondientes a la comuna de destino 'Comuna X'
comuna_especifica = df[df['Comuna Destino'] == 'Camiña']

# Mostrar todo el contenido de 'comuna_especifica'
comuna_especifica
```
