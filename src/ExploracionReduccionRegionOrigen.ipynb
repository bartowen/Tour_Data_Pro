```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/df_listo_para_segmentar.csv',sep=';')

df
```

```python
# Elimina las columnas 'Destino Turístico' y 'CUT Destino Turístico', 
df = df.drop(columns=['Destino Turistico', 'CUT Destino Turistico','Norte a Sur Origen', 'Norte a Sur Destino'])
```

```python
df.head()
```

```python
df = df.drop(columns=['Provincia Origen', 'Comuna Origen', 'CUT Comuna Origen', 'CUT Provincia Origen','CUT Provincia Destino', 'Provincia Destino'])
```

```python
# Renombrar las columnas

df = df.rename(columns={'CUT Temporada': 'Num Temporada', 'CUT Mes': 'Num Mes'})

df
```

```python
df.columns
```

```python
# Definir el nuevo orden de las columnas

column_order = ['CUT Region Origen', 'Region Origen', 'CUT Region Destino', 'Region Destino', 
                'CUT Comuna Destino', 'Comuna Destino', 'Anio', 'Num Mes', 'Mes', 
                'Num Temporada', 'Temporada', 'Viajes Ocasionales', 'PIB Region Origen', 
                'PIB Region Destino']

# Reordenar las columnas en el DataFrame

df = df[column_order]

df
```

```python
# Agrupar por las columnas
df_grouped = df.groupby(['Region Origen', 'Comuna Destino', 'Region Destino', 'Anio', 'Mes'], as_index=False).agg({
    'Viajes Ocasionales': 'sum',            # Sumar los viajes ocasionales
    'PIB Region Origen': 'first',           # Mantener el primer valor del PIB de la región de origen
    'PIB Region Destino': 'first',          # Mantener el primer valor del PIB de la región de destino
    'Temporada': 'first',                   # Mantener el primer valor de la temporada
    'CUT Region Origen': 'first',           # Mantener el primer valor de CUT Region Origen
    'CUT Comuna Destino': 'first',          # Mantener el primer valor de CUT Comuna Destino
    'CUT Region Destino': 'first',          # Mantener el primer valor de CUT Region Destino
    'Num Mes': 'first',                     # Mantener el primer valor de Num Mes
    'Num Temporada': 'first'                # Mantener el primer valor de Num Temporada
})

# Reordenar las columnas para mantener el mismo orden que el df original

df_grouped = df_grouped[['CUT Region Origen', 'Region Origen', 'CUT Region Destino', 'Region Destino', 
                        'CUT Comuna Destino', 'Comuna Destino', 'Anio', 'Num Mes', 'Mes', 
                        'Num Temporada', 'Temporada', 'Viajes Ocasionales', 
                        'PIB Region Origen', 'PIB Region Destino']]

df_grouped

```

```python
def obtener_lista_destinos(df, columna):

    lista_destinos = sorted(list(set(df[columna].unique().tolist())))
    return lista_destinos

lista_destinos = obtener_lista_destinos(df_grouped, 'Comuna Destino')
print(lista_destinos)

```

```python
#Verificar las temporadas

import seaborn as sns
import matplotlib.pyplot as plt
import math

# Visualizar temporadas por región

regiones = df['Region Destino'].unique()

# Calcular el número de filas y columnas necesario para que quede bien las regiones

n_regiones = len(regiones)
n_cols = 3
n_rows = math.ceil(n_regiones / n_cols)  # Calcular el número de filas necesario

plt.figure(figsize=(16, n_rows * 5))  # Ajustar el tamaño de la figura

# Iterar sobre cada región para crear un subplot

for i, region in enumerate(regiones, 1):
    # Crear un subplot para cada región
    plt.subplot(n_rows, n_cols, i)
    
    # Filtrar los datos por región actual

    data_region = df[df['Region Destino'] == region]
    
    # Crear el gráfico por temporada

    sns.barplot(data=data_region, x='Mes', y='Viajes Ocasionales', hue='Temporada', dodge=False,
                palette={'Alta': 'red', 'Baja': 'blue'})
    
    
    plt.title(f'Temporada Alta y Baja - {region}', fontsize=14)
    plt.xlabel('Mes', fontsize=12)
    plt.ylabel('Total de Viajes Ocasionales', fontsize=12)
    plt.xticks(rotation=45)
    plt.legend(title='Temporada')

plt.tight_layout()
plt.show()
```

```python
# revision outliers Viajes Ocasacionales

# Comparación de PIB Region Origen y PIB Regions destino con Viajes Ocasionales con respecto al Tiempo

dfp = df[['Anio','Num Mes','PIB Region Origen', 'PIB Region Destino', 'Viajes Ocasionales']]

# Asegurarnos de que 'Anio' y 'CUT Mes' existan y sean numéricos
dfp['Fecha'] = pd.to_datetime(dfp[['Anio', 'Num Mes']].astype(str).agg('-'.join, axis=1), format='%Y-%m')

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

# PIB Región Origen y Viajes Ocasionales (ax1)
sns.lineplot(data=dfp, x='Fecha', y='PIB Region Origen', ax=ax1, color='b', label='PIB Región Origen')
ax1.set_ylabel('PIB Región Origen', color='b')
ax1.tick_params(axis='y', labelcolor='b')

# Segundo eje Y para los Viajes Ocasionales 
ax2_1 = ax1.twinx()
sns.lineplot(data=dfp, x='Fecha', y='Viajes Ocasionales', ax=ax2_1, color='g', label='Viajes Ocasionales')
ax2_1.set_ylabel('Viajes Ocasionales', color='g')
ax2_1.tick_params(axis='y', labelcolor='g')

ax1.set_title("PIB de Región de Origen y Viajes Ocasionales")
ax1.set_xlabel('Fecha')

# PIB Región Destino y Viajes Ocasionales en el segundo subplot (ax2)
sns.lineplot(data=dfp, x='Fecha', y='PIB Region Destino', ax=ax2, color='r', label='PIB Región Destino')
ax2.set_ylabel('PIB Región Destino', color='r')
ax2.tick_params(axis='y', labelcolor='r')

# Segundo eje Y para los Viajes Ocasionales
ax2_2 = ax2.twinx()
sns.lineplot(data=dfp, x='Fecha', y='Viajes Ocasionales', ax=ax2_2, color='orange', label='Viajes Ocasionales')
ax2_2.set_ylabel('Viajes Ocasionales', color='orange')
ax2_2.tick_params(axis='y', labelcolor='orange')

ax2.set_title("PIB de Región de Destino y Viajes Ocasionales")
ax2.set_xlabel('Fecha')

plt.tight_layout()

plt.show()


```

```python
def plot_viajes_ocasionales_boxplot(df):

    numerical_cols = ['Viajes Ocasionales']  # Lista con la columna que deseas graficar
    
    num_plots = len(numerical_cols)
    n_cols = 2  # Número de columnas por fila
    n_rows = (num_plots + n_cols - 1) // n_cols  # Cálculo del número de filas necesarias
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(12, 3 * n_rows))
    axes = axes.flatten()

    # Graficar
    for i, col in enumerate(numerical_cols):
        sns.boxplot(y=df[col], ax=axes[i])
        axes[i].set_title(f'{col}')

    # Desactivar los ejes que no se usan
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.show()

 plot_viajes_ocasionales_boxplot(df)
```
