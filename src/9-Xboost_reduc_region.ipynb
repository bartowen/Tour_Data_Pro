```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/processed/df_segmento_Reduccion_region_origen_comuna_destino.csv',sep=';')
df
```

```python
df.shape
```

```python
df.columns
```

```python
df_num = df.drop(['Region Origen','Region Destino','Comuna Destino','Mes', 'Temporada'],axis=1)
df_num
```

```python
from sklearn.model_selection import train_test_split

X = df_num.drop('Viajes Ocasionales',axis=1)
y = df_num['Viajes Ocasionales']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

X_train
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(random_state = 42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print(f'El MAE es: {mean_absolute_error(y_test, y_pred)}')
print(f'El MSE es: {mean_squared_error(y_test, y_pred)}')
print(f'El RMSE es: {np.sqrt(mean_squared_error(y_test, y_pred))}')
print(f'El R2 es: {r2_score(y_test, y_pred)}')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [290, 310, 300],
    'max_depth': [6, 8, 7],
    'learning_rate': [0.09, 0.15, 0.1]
}

grid_search = GridSearchCV(estimator=XGBRegressor(random_state=42), param_grid=param_grid, cv=3, n_jobs=-1, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Ver los mejores parámetros encontrados
print("Mejores parámetros:", grid_search.best_params_)
```

```python
# Obtener el mejor modelo ajustado
best_model = grid_search.best_estimator_
```

```python
# Realizar predicciones en el conjunto de prueba
y_pred_best = best_model.predict(X_test)
```

```python
print(f'El MAE es: {mean_absolute_error(y_test, y_pred_best)}')
print(f'El MSE es: {mean_squared_error(y_test, y_pred_best)}')
print(f'El RMSE es: {np.sqrt(mean_squared_error(y_test, y_pred_best))}')
print(f'El R2 es: {r2_score(y_test, y_pred_best)}')
```
