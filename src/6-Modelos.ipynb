```python
import pandas as pd
```

```python
df = pd.read_csv('../data/processed/df_listo_para_eda.csv',sep=';')
df.head()
```

```python
df.shape
```

```python
# Factorizar las variables cátegoricas, para poder hacer matriz de correlaciónes 
df['CUT Destino Turistico'] = pd.factorize(df['Destino Turistico'])[0]
df['CUT Temporada'] = pd.factorize(df['Temporada'])[0]

# Ordenar columnas
columnas = [
    'CUT Comuna Origen', 'Comuna Origen', 'CUT Provincia Origen', 'Provincia Origen',
    'CUT Region Origen', 'Region Origen', 'CUT Comuna Destino', 'Comuna Destino', 'CUT Destino Turistico',
    'Destino Turistico', 'CUT Provincia Destino', 'Provincia Destino', 'CUT Region Destino',
    'Region Destino', 'Anio', 'CUT Mes', 'Mes', 'Viajes Ocasionales', 'CUT Temporada', 'Temporada',
    'PIB Region Origen', 'PIB Region Destino', 'Total Temporada Region', 'Total Temporada Provincia', 'Total Temporada Comuna'
]

df = df[columnas]
```

```python
df = df.drop(['Total Temporada Region','Total Temporada Provincia','Total Temporada Comuna'],axis=1)
df.head()
```

```python
df_num = df.drop(['Comuna Origen', 'Provincia Origen', 'Region Origen', 'Comuna Destino', 'Destino Turistico', 'Provincia Destino', 'Region Destino', 'Mes', 'Temporada'],axis=1)
df_num
```

```python
from sklearn.model_selection import train_test_split

X = df_num.drop('Viajes Ocasionales',axis=1)
y = df_num['Viajes Ocasionales']

X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, random_state=42)

X_train
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=42)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)

from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print(f'El MAE es: {mean_absolute_error(y_pred,y_test)}')
print(f'El MSE es: {mean_squared_error(y_pred,y_test)}')
print(f'El R2 es: {r2_score(y_pred,y_test)}')
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler

# Selección de características con SelectKBest (Ensayo y Error)
selector = SelectKBest(score_func=f_regression, k=8)
X_train_sel = selector.fit_transform(X_train, y_train)

# Transformación a los datos de test
X_test_sel = selector.transform(X_test)

# Obtener las columnas seleccionadas
selected_features = selector.get_support(indices=True)
best_columns = X.columns[selected_features]
print("Columnas seleccionadas:", best_columns)
```

```python
# Escalar las características seleccionadas con MinMaxScaler
scaler = MinMaxScaler()
X_train_esc = scaler.fit_transform(X_train_sel) 
X_test_esc = scaler.transform(X_test_sel)  

model = RandomForestRegressor(random_state=42)
model.fit(X_train_esc,y_train)

y_pred = model.predict(X_test_esc)

print(f'El MAE es: {mean_absolute_error(y_pred,y_test)}')
print(f'El MSE es: {mean_squared_error(y_pred,y_test)}')
print(f'El R2 es: {r2_score(y_pred,y_test)}')
```

```python
# Escalar las características seleccionadas con StandardScaler
scaler = StandardScaler()
X_train_esc = scaler.fit_transform(X_train_sel) 
X_test_esc = scaler.transform(X_test_sel)  

model = RandomForestRegressor(random_state=42)
model.fit(X_train_esc,y_train)

y_pred = model.predict(X_test_esc)

print(f'El MAE es: {mean_absolute_error(y_pred,y_test)}')
print(f'El MSE es: {mean_squared_error(y_pred,y_test)}')
print(f'El R2 es: {r2_score(y_pred,y_test)}')
```

```python
# Escalar las características seleccionadas con RobustScaler
scaler = RobustScaler()
X_train_esc = scaler.fit_transform(X_train_sel) 
X_test_esc = scaler.transform(X_test_sel)  

model = RandomForestRegressor(random_state=42)
model.fit(X_train_esc,y_train)

y_pred = model.predict(X_test_esc)

print(f'El MAE es: {mean_absolute_error(y_pred,y_test)}')
print(f'El MSE es: {mean_squared_error(y_pred,y_test)}')
print(f'El R2 es: {r2_score(y_pred,y_test)}')
```
