```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/processed/df_listo_para_segmentar.csv',sep=';')

df
```

```python
# Elimina las columnas 'Destino Turístico' y 'CUT Destino Turístico', 
df = df.drop(columns=['CUT Destino Turistico','Destino Turistico','Norte a Sur Origen', 'Norte a Sur Destino'])
```

```python
df.columns
```

```python
total_combinaciones_viajes = df.groupby(['CUT Comuna Origen', 'Comuna Origen', 'CUT Provincia Origen',
       'Provincia Origen', 'CUT Region Origen', 'Region Origen',
       'CUT Comuna Destino', 'Comuna Destino', 'CUT Provincia Destino',
       'Provincia Destino', 'CUT Region Destino', 'Region Destino', 'Anio',
       'CUT Mes', 'Mes', 'CUT Temporada', 'Temporada',
       'PIB Region Origen', 'PIB Region Destino']).sum().reset_index()
```

```python
# Crear columna 'covid_periodo' con valores 0 o 1
df['covid_periodo_num'] = df['Anio'].apply(lambda x: 1 if x in [2020, 2021] else 0)

# Crear columna 'covid_periodo_texto' con valores 'Sí' o 'No'
df['covid_periodo'] = df['Anio'].apply(lambda x: 'Sí' if x in [2020, 2021] else 'No')

df
```

```python
df_num = df.drop(['Comuna Origen', 'Provincia Origen', 'Region Origen', 'Comuna Destino','Provincia Destino', 'Region Destino', 'Mes', 'Temporada','covid_periodo'],axis=1)
df_num
```

```python
target_variable = 'Viajes Ocasionales'

# Crear x_train y x_test excluyendo la columna objetivo

X_train = df_num[df['Anio'] != 2024].drop(columns=[target_variable])
X_test = df_num[df_num['Anio'] == 2024].drop(columns=[target_variable])

# Crear y_train y y_test con solo la columna objetivo

y_train = df_num[df_num['Anio'] != 2024][target_variable]
y_test = df_num[df_num['Anio'] == 2024][target_variable]

X_test
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(random_state = 42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print(f'El MAE es: {mean_absolute_error(y_test, y_pred)}')
print(f'El MSE es: {mean_squared_error(y_test, y_pred)}')
print(f'El RMSE es: {np.sqrt(mean_squared_error(y_test, y_pred))}')
print(f'El R2 es: {r2_score(y_test, y_pred)}')
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, root_mean_squared_error

# Selección de características con SelectKBest (Ensayo y Error)
selector = SelectKBest(score_func=f_regression, k=11)
X_train_sel = selector.fit_transform(X_train, y_train)

# Transformación a los datos de test
X_test_sel = selector.transform(X_test)

# Obtener las columnas seleccionadas
selected_features = selector.get_support(indices=True)
best_columns = X_train.columns[selected_features]
print("Columnas seleccionadas:", best_columns)
```

```python
best_model = XGBRegressor(random_state = 42)
best_model.fit(X_train_sel, y_train)
```

```python
y_pred_best = best_model.predict(X_test_sel)
y_pred_best
```

```python
print(f'El MAE es: {mean_absolute_error(y_test, y_pred_best)}')
print(f'El MSE es: {mean_squared_error(y_test, y_pred_best)}')
print(f'El RMSE es: {np.sqrt(mean_squared_error(y_test, y_pred_best))}')
print(f'El R2 es: {r2_score(y_test, y_pred_best)}')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators': [290, 310, 300],
    'max_depth': [6, 8, 7],
    'learning_rate': [0.09, 0.15, 0.1]
}

grid_search = GridSearchCV(estimator=XGBRegressor(random_state=42), param_grid=param_grid, cv=3, n_jobs=-1, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Ver los mejores parámetros encontrados
print("Mejores parámetros:", grid_search.best_params_)
```

```python
# Obtener el mejor modelo ajustado
best_model = grid_search.best_estimator_
```

```python
# Realizar predicciones en el conjunto de prueba
y_pred_best = best_model.predict(X_test)
```

```python
print(f'El MAE es: {mean_absolute_error(y_test, y_pred_best)}')
print(f'El MSE es: {mean_squared_error(y_test, y_pred_best)}')
print(f'El RMSE es: {np.sqrt(mean_squared_error(y_test, y_pred_best))}')
print(f'El R2 es: {r2_score(y_test, y_pred_best)}')
```
