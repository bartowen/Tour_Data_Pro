```python
import pandas as pd
from scipy import stats
```

```python
df = pd.read_csv('../data/processed/df_listo_para_eda.csv', sep=';')
df
```

HIPÓTESIS 1:

    Influencia de la Temporada en los viajes ocasionales dentro de Chile.

            Hipótesis nula: "La cantidad de viajes ocasionales no difiere entre la temporada alta y la temporada baja, independientemente del año."
            Hipótesis alternativa:  "La cantidad de viajes ocasionales difiere entre la temporada alta y la temporada baja, independientemente del año."

```python

# Separar los datos por temporada
alta = df[df['Temporada'] == 'Alta']['Viajes Ocasionales']
baja = df[df['Temporada'] == 'Baja']['Viajes Ocasionales']

# Realizar el t-test
t_stat, p_value = stats.ttest_ind(alta, baja)

# Mostrar resultados
print(f"Estadístico t: {t_stat}, Valor p: {p_value}")

# Interpretar el resultado
alpha = 0.05
if p_value <= alpha:
    print("Rechazamos la hipótesis nula. Hay una diferencia significativa en la cantidad de viajes.")
else:
    print("No podemos rechazar la hipótesis nula. No hay evidencia de una diferencia significativa.")
```

HIPÓTESIS 2:


        Efecto del PIB en las Predicciones de Viajes


                Hipótesis Nula: "El PIB de la región de origen no tiene un efecto significativo en la cantidad de viajes ocasionales realizados."

                Hipótesis Alternativa: "El PIB de la región de origen tiene un efecto significativo en la cantidad de viajes ocasionales realizados."
E

```python
import statsmodels.api as sm

# Definir las variables dependientes e independientes
X = df['PIB Region Origen']  # Variable independiente (PIB)
y = df['Viajes Ocasionales']  # Variable dependiente (Viajes)

# Agregar una constante a X para el término independiente en el modelo
X = sm.add_constant(X)

# Ajustar el modelo de regresión lineal
modelo = sm.OLS(y, X).fit()

# Resumen de resultados
print(modelo.summary())

# Interpretación de los resultados
p_value = modelo.pvalues[1]  # Obtener el p-value del coeficiente de PIB

# Comparar con un nivel de significancia
alpha = 0.05
if p_value < alpha:
    print("\n*** Rechazamos la hipótesis nula: El PIB de la región de origen tiene un efecto significativo en la cantidad de viajes. ***\n")
else:
    print("\n*** No rechazamos la hipótesis nula: El PIB de la región de origen no tiene un efecto significativo en la cantidad de viajes. ***\n")
```
