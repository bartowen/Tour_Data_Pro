```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/processed/df_listo_para_segmentar.csv',sep=';')

df
```

```python
# Elimina las columnas 'Destino Turístico' y 'CUT Destino Turístico', 
df = df.drop(columns=['Norte a Sur Origen', 'Norte a Sur Destino'])
```

```python
df.columns
```

```python
# Crear columna 'covid_periodo' con valores 0 o 1
df['covid_periodo_num'] = df['Anio'].apply(lambda x: 1 if x in [2020, 2021] else 0)

# Crear columna 'covid_periodo_texto' con valores 'Sí' o 'No'
df['covid_periodo'] = df['Anio'].apply(lambda x: 'Sí' if x in [2020, 2021] else 'No')

df
```

```python
df_num = df.drop(['Comuna Origen', 'Provincia Origen', 'Region Origen', 'Comuna Destino', 'Destino Turistico','Provincia Destino', 'Region Destino', 'Mes', 'Temporada','covid_periodo'],axis=1)
df_num
```

```python
target_variable = 'Viajes Ocasionales'

# Crear x_train y x_test excluyendo la columna objetivo

X_train = df_num[df['Anio'] != 2023].drop(columns=[target_variable])
X_test = df_num[df_num['Anio'] == 2023].drop(columns=[target_variable])

# Crear y_train y y_test con solo la columna objetivo

y_train = df_num[df_num['Anio'] != 2023][target_variable]
y_test = df_num[df_num['Anio'] == 2023][target_variable]

X_test
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(random_state = 42)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

print(f'El MAE es: {mean_absolute_error(y_test, y_pred)}')
print(f'El MSE es: {mean_squared_error(y_test, y_pred)}')
print(f'El RMSE es: {np.sqrt(mean_squared_error(y_test, y_pred))}')
print(f'El R2 es: {r2_score(y_test, y_pred)}')
```

```python
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error, root_mean_squared_error

# Selección de características con SelectKBest
selector = SelectKBest(score_func=f_regression, k=8)
X_train_sel = selector.fit_transform(X_train, y_train)

# Transformación de los datos de test con las mismas características seleccionadas
X_test_sel = selector.transform(X_test)

# Obtener las columnas seleccionadas
selected_features = selector.get_support(indices=True)
best_columns = X_train.columns[selected_features]
print("Columnas seleccionadas:", best_columns)

# Ahora, puedes proceder a entrenar el modelo con las características seleccionadas
# Normalización de los datos usando MinMaxScaler (opcional, dependiendo de tu modelo)
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train_sel)
X_test_scaled = scaler.transform(X_test_sel)

# Crear el modelo de Random Forest
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train_scaled, y_train)

# Realizar predicciones
y_pred = model.predict(X_test_scaled)

# Evaluar el modelo
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R2: {r2}")
```
